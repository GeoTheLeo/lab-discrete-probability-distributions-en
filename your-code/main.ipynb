## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
from scipy.stats import bernoulli

# Probability of apple (success)
p_apple = 60/100

# Create a Bernoulli distribution with p = p_apple
bern = bernoulli(p_apple)

# Probability of apple (success): P(X=1)
prob_apple = bern.pmf(1)

# Probability of orange (failure): P(X=0)
prob_orange = bern.pmf(0)

print(f"1. Probability that the fruit is an apple: {prob_apple:.2f}")
print(f"2. Probability that the fruit is an orange: {prob_orange:.2f}")
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
# What is the probability that the first 5 fruits are all apples?
from scipy.stats import bernoulli

p = 0.6  # probability of apple
q = 0.4  # probability of orange

# 1. Probability that first 5 are apples
prob_5_apples = p ** 5

# 2. Probability that first 5 are apples and next 15 are oranges
prob_5_apples_15_oranges = (p ** 5) * (q ** 15)

print(f"1. Probability that the first 5 fruits are all apples: {prob_5_apples:.6f}")
print(f"2. Probability that the first 5 are apples and the next 15 are oranges: {prob_5_apples_15_oranges:.10f}")
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# your solution here
from scipy.stats import binom

n = 20   # number of trials
p = 0.6  # probability of apple
k = 5    # number of apples desired

# Probability of exactly 5 apples
prob_5_apples = binom.pmf(k, n, p)

print(f"Probability that the sample contains precisely 5 apples and 15 oranges: {prob_5_apples:.10f}")
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# your code here
from scipy.stats import binom

n = 20   # number of trials
p = 0.6  # probability of apple
k = 4    # we want P(X <= 4)

# Cumulative probability: P(X <= 4)
prob_less_than_5 = binom.cdf(k, n, p)

print(f"Probability that less than 5 fruits are apples (i.e., 0 to 4 apples): {prob_less_than_5:.10f}")
````

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 20   # number of trials
p = 0.6  # probability of apple

# Range of k values: from 0 to n
k_values = np.arange(0, n+1)

# PMF for each k
pmf_values = binom.pmf(k_values, n, p)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution PMF (n=20, p=0.6)')
plt.xlabel('Number of Apples (k)')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
# your code here 
import math

lambda_ = 2.3  # average goals per match
k = 5          # number of goals

# Calculate P(X = 5)
numerator = (lambda_ ** k) * math.exp(-lambda_)
denominator = math.factorial(k)
probability = numerator / denominator

print(f"Probability that exactly 5 goals are scored: {probability:.4f}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
import numpy as np
import matplotlib.pyplot as plt
import math

lambda_ = 2.3  # average goals per match
k_values = np.arange(0, 11)  # goals from 0 to 10

# Calculate the PMF for each k
pmf_values = [ (lambda_**k * math.exp(-lambda_)) / math.factorial(k) for k in k_values ]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title('Poisson Distribution PMF (=2.3) - Goals in a Match')
plt.xlabel('Number of Goals')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````
